```python
import numpy as np
import pandas as pd # to read dataset
import warnings # to remove warnings
warnings.filterwarnings('ignore')
```

# Loading first dataset


```python
first_dataset = pd.read_csv('/content/dataset.csv')
first_dataset
```

```python
second_dataset = pd.read_csv('/content/IPL Matches 2008-2020.csv')
second_dataset
```

# Data cleaning


```python
# finding total runs scored in each innings
total_runs = first_dataset.groupby(['match_id','inning']).sum()['total_runs'].add(1).reset_index()
total_runs
```

```python
# selecting only first innings score to find target of each match
total_runs = total_runs[total_runs['inning'] == 1]
total_runs
```

```python
# merging the total row columns based on match_id
match_df = second_dataset.merge(total_runs[['match_id','total_runs']],left_on='id',right_on='match_id')
```

```python
match_df
```

```python
# for final dataset selecting only these columns
match_df = match_df[['id','city','winner','total_runs']]
```

```python
match_df = match_df.rename(columns={'id':'match_id'})
```

```python
second_innings = match_df.merge(first_dataset, on='match_id')
second_innings
```

```python
# here we will be predicting the winner shape is 86240 and 24 col
# but 24 col are to much we only want 9 or 10 rows
second_innings  = second_innings[second_innings['inning'] == 2]
second_innings.shape
```

```python
second_innings['batting_team'].value_counts()
```

```python
second_innings['batting_team'] = second_innings['batting_team'].replace('Delhi Daredevils','Delhi Capitals')
second_innings['bowling_team'] = second_innings['bowling_team'].replace('Delhi Daredevils','Delhi Capitals')

second_innings['batting_team'] = second_innings['batting_team'].replace('Deccan Chargers','Sunrisers Hyderabad')
second_innings['bowling_team'] = second_innings['bowling_team'].replace('Deccan Chargers','Sunrisers Hyderabad')
```

```python
second_innings['batting_team'].value_counts()
```

```python
second_innings['bowling_team'].value_counts()
```

```python
teams = ['Kolkata Knight Riders',
         'Chennai Super Kings',
         'Kings XI Punjab',
         'Rajasthan Royals',
         'Mumbai Indians',
         'Delhi Capitals',
         'Royal Challengers Bangalore',
         'Sunrisers Hyderabad']
```

```python
second_innings = second_innings[second_innings['batting_team'].isin(teams)]
second_innings = second_innings[second_innings['bowling_team'].isin(teams)]
# remove all data except these 8 teams using isin function
# so isin func simply check that the team name among this list or not if
# it is not in that list it will simply exclude that row and if it is in the
# list it will consider the row
```

now we will find the current score of the dataset but in dataset there is no such column present but using the column
total_runs_y we can find the current score using the cumulative solve function

this column store runs made by batsman at particular at particular ball so to calulate after each ball we will use a concept called as cumulative sum we just go on adding adding and after at the end we just get the total sum of column so we will use the same concept here.

```python
# Finding current score using cumulative sum function
second_innings['current_Score'] = second_innings.groupby('match_id')['total_runs_y'].cumsum()
second_innings
# This approach is applied after the groupby operation to calculate the cumulative sum specifically for that column.
```

```python
 # calculation of runs left
 second_innings['runs_left'] = second_innings['total_runs_x'] - second_innings['current_Score']
 second_innings
 # total_run = 208 - currentscore(1) = run_left = 207
```

```python
#calulating ball left after each ball is thrown
second_innings['balls_left'] = 126 - (second_innings['over']*6 + second_innings['ball'])
second_innings
# dervied from the total no of balls in innings - no of balls already played
```

# Getting wickets

```python
#player_dismissed tell if batsmen was out or not while playing that ball
#fill null value with 0
second_innings['player_dismissed'] = second_innings['player_dismissed'].fillna("0")

#if column value is not zero mark it 1
second_innings['player_dismissed'] = second_innings['player_dismissed'].apply(lambda x:x if x == "0" else "1")

#changing datatype of column  name
second_innings['player_dismissed'] = second_innings['player_dismissed'].astype('int')

#finding wickets using cumulative sum function
wickets = second_innings.groupby('match_id')['player_dismissed'].cumsum()

#calculating wickets left by simple maths operations
second_innings['wickets'] = 10 - wickets
second_innings

#

#
```

# Current Runrate


```python
#CRR formula
second_innings['crr'] = (second_innings['current_Score']*6)/(120 - second_innings['balls_left'])
second_innings['crr'] = second_innings['crr'].round(2)  # Round to 2 decimal places
second_innings
```

# Required Runrate


```python
# rrr formula
second_innings['rrr'] = round((second_innings['runs_left']*6)/(second_innings['balls_left']),2)
second_innings
# runsleft / overleft
# using round func rounded the answer upto 2 decimal places
```

```python
# function to mark winner team 0 or 1 based on winner column
def winner(team):
    return 1 if team['batting_team'] == team['winner'] else 0
```

```python
second_innings['results'] = second_innings.apply(winner,axis=1)
second_innings.sample(10)
```

# Only taking required columns

```python
required_data = second_innings[['batting_team','bowling_team','city','runs_left','balls_left','wickets','total_runs_x','crr','rrr','results']]
required_data
```

# Sampling Data

```python
#sampling dataset for proper model building so that our model will be more
# accurate
final_data = required_data.sample(required_data.shape[0])
```

# Dropping NULL values

```python
final_data.dropna(inplace=True)
```

# Removing rows where ball_left=0 to reomve -inf and inf from rrr column

```python
# tha value in minus not suitable for model building
# it simply remove the rows from our final dataset where the ball left is zero
final_data = final_data[final_data['balls_left'] != 0]
```

# Storing data in new csv

```python
final_data.to_csv('final_data.csv',index=False)
```

# Building Model

```python
from sklearn.model_selection import train_test_split as tts
```

```python
x = final_data.iloc[:, :-1]
y = final_data.iloc[:,-1]
# result column in y variable
# remainig all column in X variable
```

```python
x_train,x_test,y_train,y_test = tts(x,y,test_size=0.25,random_state=100)
x_train
# train , test and split
```

```python
# test_data
x_test
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
# using oneHotEncoder for making textual columns to numerical column for model
# building
# The 'sparse' argument has been replaced with 'sparse_output' in newer versions of scikit-learn.
trf = ColumnTransformer([
    ('trf',OneHotEncoder(sparse_output=False,drop='first'),['batting_team','bowling_team','city'])
],
remainder='passthrough') #passthrough confirms that no other column were touch
# during this process
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
# creating pipeline (building model)
pipe = pipe = Pipeline(steps=[
    ('step1',trf),
    ('step2',LogisticRegression(solver='liblinear'))
])
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred = pipe.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

# Random Forest classifier



```python
from sklearn.ensemble import RandomForestClassifier
trf = ColumnTransformer([
    ('trf',OneHotEncoder(sparse_output=False,drop='first'),['batting_team','bowling_team','city']) # Change 'sparse' to 'sparse_output'
],
remainder='passthrough')

pipe = Pipeline([
    ('step1',trf),
    ('step2',RandomForestClassifier())
])

pipe.fit(x_train,y_train)

y_pred = pipe.predict(x_test)

accuracy_score(y_test,y_pred)
```
